###### 2.判断某一年为闰年

```python
year =int(input("请输入年份： ")) 

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("{0}是闰年".format(year))
        else:
            print("{0}不是闰年".format(year))
    else:
        print("{0}是闰年".format(year))        
else:
    print("{0}不是闰年！".format(year))
    
if year % 4 == 0 and year % 100 != 0 or  year % 400 == 0:
    print("{0}是闰年".format(year))
else:
    print("0不是闰年".format(year))
```

```python
3.ATM机功能

查询余额 存款 取款 退出
```

```python
windows = """
            ===========
            1.查询余额
            2.存款
            3.取款
            4.退出
            ===========
          """
balance = 100  
expense = 0

text=int(input(windows))

while text != 4:
    text=int(input(windows))
    if text == 1:
        print("余额为:",balance)
    elif text == 2:
        balance  + int(input("存款金额为："))
    else:
        expense = int(input("取款金额为："))
        if balance - expense < 0:
            print("余额不足")
        else:
            balance = balance - expense
                  
else:
        print("退出")
```
